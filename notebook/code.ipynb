```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
x_train.corr()
```

```python
plt.figure(figsize=(12,10))
corr=x_train.corr()
sns.heatmap(corr,annot=True)
```

```python
def correlation(dataset,threshold):
    col_corr=set() 
    corr_mat=dataset.corr()
    for i in range(len(corr_mat.columns)):
        for j in range (i):
            if abs(corr_mat.iloc[i,j])>threshold:
                colname=corr_mat.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(x_train,0.85)
```

```python
corr_features
```

```python
x_train.drop(corr_features,axis=1,inplace=True)
x_test.drop(corr_features,axis=1,inplace=True)
x_train.shape,x_test.shape
```

Standardisation

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled
```

box plot to understand effect of standard scaler 

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=x_train)
plt.title("X train before scaling")
plt.subplot(1,2,2)
sns.boxplot(data=x_train_scaled)
plt.title("X train after scaling")
plt.show()
```

```python
from sklearn.linear_model import LassoCV
lasocv=LassoCV(cv=5)
lasocv.fit(x_train_scaled,y_train)
```

```python
lasocv.predict(x_test_scaled)
```

```python
lasocv.alpha_
```

```python
lasocv.mse_path_
```

```python
from sklearn.metrics import mean_absolute_error,r2_score

y_pred=lasocv.predict(x_test_scaled)
plt.scatter(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print("mean abso;ute error ",mae)
print("r2 score ", score)
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score 

linaerR=LinearRegression()
linaerR.fit(x_train_scaled,y_train)
y_pred=linaerR.predict(x_test_scaled)

mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print(" The mean absolute error ",mae)
print("The r2 score",score)

plt.scatter(y_test,y_pred)
```

```python
## Lasso Regression model 
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score 

lso=Lasso()
lso.fit(x_train_scaled,y_train)
y_pred=lso.predict(x_test_scaled)

mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print(" The mean absolute error ",mae)
print("The r2 score",score)

plt.scatter(y_test,y_pred)
```

```python
##Ridge Regression Model 

from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score 

rdg=Ridge()
rdg.fit(x_train_scaled,y_train)
y_pred=rdg.predict(x_test_scaled)

mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print(" The mean absolute error ",mae)
print("The r2 score",score)

plt.scatter(y_test,y_pred)
```

```python
## ElasticNet 

from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score 

elnet=ElasticNet()
elnet.fit(x_train_scaled,y_train)
y_pred=elnet.predict(x_test_scaled)

mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)

print(" The mean absolute error ",mae)
print("The r2 score",score)

plt.scatter(y_test,y_pred)
```
